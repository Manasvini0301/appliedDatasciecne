```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
food = pd.read_csv('WorldFoodProduction.csv')

# Select data for the five countries
countries = ['India', 'China', 'St. Kitts and Nevis', 'Albania', 'Puerto Rico','Pakistan','St. Lucia']

food_selected = food[food['Country Name'].isin(countries)]

# Plot the line chart
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = food_selected[food_selected['Country Name'] == country]
    plt.plot(country_data.columns[4:], country_data.iloc[0, 4:].values, label=country)

plt.xlabel('Year')
plt.ylabel('Food Production Index')
plt.title('Food Production Index for Selected Countries')

# Adjusting x-axis labels to show every 5 years
plt.xticks(list(food.columns[4::5]))

plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = food_selected[food_selected['Country Name'] == country]
    plt.bar(country_data.columns[4:], country_data.iloc[0, 4:].values, label=country)

plt.xlabel('Country')
plt.ylabel('Food Production Index')
plt.title('Food Production Index for Selected Countries in ')
plt.xticks(list(food.columns[4::5]))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
boxplot_data = [food_selected[food_selected['Country Name'] == country].iloc[0, 4:].values for country in countries]
plt.boxplot(boxplot_data, labels=countries)
plt.xlabel('Country')
plt.ylabel('Food Production Index')
plt.title('Box Plot of Food Production Index for Selected Countries')
plt.xticks(rotation=90)
plt.show()
```
